## Simple simulation with California data

```python
import covasim as cv
cv.options(jupyter=True, verbose=0)

pars = dict(
    start_day = '2021-01-01',
    end_day   = '2021-03-01',
    beta      = 0.3,
    pop_size = 40000,  # California population is 40 million
    pop_type  = 'hybrid'
)
sim = cv.Sim(pars=pars, datafile='california_data/covid_CA_jan_feb.csv', interventions=cv.test_num(daily_tests='data'))
sim.run()
sim.plot(['cum_tests', 'cum_diagnoses', 'cum_deaths'])
```

## Adding Interventions - Vaccinations

```python
import sciris as sc
import covasim as cv

# Create simulation
pars = sc.objdict(
    pop_size       = 40_000,
    start_day = '2021-01-01',
    end_day   = '2021-03-01',
    beta           = 0.015,
    rel_death_prob = 0.7,
    interventions  = cv.test_num(daily_tests='data'), # adding number of daily test as an intervention
    verbose        = 0,
    pop_type  = 'hybrid',
    rand_seed = 96, # Random seed for reproducibility
)

sim = cv.Sim(pars, datafile='covid_CA_jan_feb.csv')

calib_pars = dict(
    beta      = [0.013, 0.005, 0.020],
    test_prob = [0.01, 0.00, 0.30]
)

def set_test_prob(sim, calib_pars):
    tp = sim.get_intervention(cv.test_prob)
    tp.symp_prob = calib_pars['test_prob']
    return sim

calib = sim.calibrate(calib_pars=calib_pars, custom_fn=set_test_prob, total_trials=1)
calib.plot_all()
calib.plot_sims(to_plot=['cum_tests', 'cum_diagnoses', 'cum_deaths','n_susceptible'])
```

```python
# For Covasim’s vaccination
sim['interventions'] = [cv.vaccinate_prob(vaccine='pfizer', # Type of vaccination
                                          days=20, # Day(s) to apply intervention
                                          prob=0.8)] # Probability of vaccination on this day
# For pharmacy based vaccination
sim['interventions'] = [cv.vaccinate_geo(
    vaccine='pfizer', # Type of vaccination
    days=20, # Day(s) to apply intervention
    prob=0.8)] # Base probability of vaccination on this day
sim['pharmacies'] = [cv.Pharmacy(name='Vaccines',coords={'lat': 0, 'lon': 0},offer_vaccines=True)]
```

## Testing Interventions California data

```python
import covasim as cv
import matplotlib.pyplot as plt
# Define the testing interventions
tn_data = cv.test_num('data')
#tn_fixed = cv.test_num(daily_tests=500, start_day='2021-01-01')
tp = cv.test_prob(symp_prob=0.9, asymp_prob=0.001, start_day='2021-01-01')

# Define the default parameters
pars = dict(
        pop_size       = 400000,
        start_day = '2021-01-01',
        end_day   = '2021-03-01',
        beta           = 1.9,
        rel_death_prob = 0.7,
        #interventions  = cv.test_num(daily_tests='data'), # adding number of daily test as an intervention
        verbose        = 0,
        pop_type  = 'hybrid',
        rand_seed = 96, # Random seed for reproducibility
        pop_infected = 200
)




# Define the simulations
sim1 = cv.Sim(pars, datafile='covid_CA_jan_feb.csv', interventions=tn_data, label='Number of tests from data')
#sim2 = cv.Sim(pars, interventions=tn_fixed, label='Constant daily number of tests')
sim3 = cv.Sim(pars, interventions=tp, label='Probability-based testing')

# Run and plot results
#msim = cv.parallel(sim1, sim2, sim3)
msim = cv.parallel(sim1,sim3)
msim.plot(['new_infections'])
```

```python
# Let's try plotting real data 
plt.figure(figsize = (27,10))
plt.rcParams.update({'font.size': 16})
plt.xticks(rotation = 75)
plt.title("New infections real data_California from 01/01/21 to 03/01/21") 
plt.plot(cal_df['date'], cal_df['new_diagnoses'], label='Real infections', marker='o', linestyle='-', color='b')
```
