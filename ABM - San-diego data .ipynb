```python
df_San_Diego = df_San_Diego.loc[df_San_Diego['county'] == 'San Diego'] # Scrapping only Alameda county data
df_San_Diego = df_San_Diego.rename(columns={'total_tests': 'new_tests', 'positive_tests': 'new_diagnoses','deaths': 'new_deaths' })
df_San_Diego['date'] = pd.to_datetime(df_San_Diego['date'])
df_San_Diego
```

# # Simple simulation

```python
import covasim as cv
cv.options(jupyter=True, verbose=0)

pars = dict(
    pop_size  = 3370418,
    start_day = '2020-12-01',
    end_day   = '2021-03-01',
    beta = 5
)
sim = cv.Sim(pars=pars, datafile='covid_county_df_San_Diego.csv')
sim.run()
sim.plot(to_plot=['new_infections', 'cum_diagnoses', 'cum_deaths','cum_tests'])
```

```python
import covasim as cv
cv.options(jupyter=True, verbose=0)

pars = dict(
    verbose = 0,
    pop_size  = 337041,
    start_day = '2020-12-01',
    end_day   = '2021-03-01',
    interventions = cv.test_prob(symp_prob=0.01),
)

sim = cv.Sim(pars, datafile='covid_county_df_San_Diego.csv')

calib_pars = dict(
    beta      = [0.013, 0.005, 0.020],
    test_prob = [0.01, 0.00, 0.30]
)

def set_test_prob(sim, calib_pars):
    tp = sim.get_intervention(cv.test_prob)
    tp.symp_prob = calib_pars['test_prob']
    return sim

calib = sim.calibrate(calib_pars=calib_pars, custom_fn=set_test_prob, total_trials=60)
calib.plot_all()
calib.plot_sims(to_plot=['new_infections', 'cum_diagnoses', 'cum_deaths','cum_tests'])

```

```python
import sciris as sc
import covasim as cv

# Create default simulation
pars = sc.objdict(
    pop_size  = 337041,
    start_day = '2020-12-01',
    end_day   = '2021-03-01',
    beta           = 0.2,
    rel_death_prob = 1.6,
    interventions  = cv.test_num(daily_tests='data'),
    verbose        = 0,
)
sim = cv.Sim(pars, datafile='covid_county_df_San_Diego.csv')

# Parameters to calibrate -- format is best, low, high
calib_pars = dict(
    beta           = [pars.beta, 0.005, 0.20],
    rel_death_prob = [pars.rel_death_prob, 0.5, 3.0],
)

if __name__ == '__main__':

    # Run the calibration
    calib = sim.calibrate(calib_pars=calib_pars, total_trials=100)
```

```python
calib.plot_sims(to_plot=['new_infections', 'cum_diagnoses', 'cum_deaths','cum_tests'])
```
